Boolean:
    1 -> True
    0 -> False
    [x,y]
    and -> if x is false then x,else y 
    or -> if x is true then x,else y
    not -> true to false and vice versa

```python
exp1 = 1 == 2
print("exp1 is ",exp1) 
exp2 = 7 > 3
print("exp2 is",exp2)
#and
print("and operaion between exp1 and exp2",exp1 and exp2)   #x and y ,x is false 
#or
print("or operation between exp1 and exp2",exp1 or exp2)    #x or y ,x is false
exp3 = 3 < 1
print("exp3 is",exp3)
print("or operation between exp1 and exp3",exp1 or exp3)    #x or y ,x is false
#not
print("not operation on exp1",not exp1)
print("not operation on exp2",not exp2)
print("not operation on exp3",not exp3)
```
